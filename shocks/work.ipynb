```python
import os
import sys
```

```python
home = os.environ['HOME']
work_path = os.path.join(home, 'far/farMe/shocks')
```

```python
import pyphare
import matplotlib.pyplot as plt
from pyphare.pharesee.run import Run
import numpy as np
```

```python
run_name = "run"
```

```python
run_path = os.path.join(work_path, run_name)
```

```python
run  = Run(run_path)
```

```python
time=0
r = Run(run_path)
back = r.GetParticles(time, ["left",])

fig, ax = plt.subplots(figsize=(10,4))

p,(f, a) = back.dist_plot(axis=("x", "Vx"),
                          ax=ax,
                          norm=0.4,
                          finest=True,
                          cmap="viridis",
                          gaussian_filter_sigma=1,
                          vmin=-1,vmax=1,
                          dv=0.02,
                          xlabel="$X \\ (d_p)$",
                          ylabel="$V_x \\ (A_0)$",
                          title="weak perturbation at time : {:.2f}".format(time),
                    )
```

```python
time=0
r = Run(run_path)
back = r.GetParticles(time, ["back",])
left = r.GetParticles(time, ["left",])
right = r.GetParticles(time, ["right",])

fig, ax = plt.subplots(figsize=(10,4))

p,(f, a) = back.dist_plot(axis=("x", "Vx"),
                          ax=ax,
                          finest=True,
                          plain=True,
                          stride=1,
                          color='k',
                          alpha=1,
                          markersize=1,
                          vmin=-1,vmax=1,
                          title="weak perturbation at time : {:.2f}".format(time),
                         )

p,(f, a) = left.dist_plot(axis=("x", "Vx"),
                          ax=ax,
                          finest=True,
                          plain=True,
                          stride=1,
                          color='r',
                          alpha=1,
                          markersize=1,
                          vmin=-1,vmax=1,
                          title="weak perturbation at time : {:.2f}".format(time),
                         )

p,(f, a) = right.dist_plot(axis=("x", "Vx"),
                          ax=ax,
                          finest=True,
                          plain=True,
                          stride=1,
                          color='b',
                          alpha=1,
                          markersize=1,
                          vmin=-1,vmax=1,
                          title="weak perturbation at time : {:.2f}".format(time),
                         )
```
