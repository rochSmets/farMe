```python
import sys
sys.path.append("/home/smets/far/PHARE/pyphare/")
import os
import subprocess
import pyphare
import matplotlib.pyplot as plt
from pyphare.pharesee.run import Run
from pyphare.core.operators import dot, cross, sqrt, modulus, grad
import numpy as np
```

```python
run_path = "/home/smets/far/farMe/ionBeam/yao/run/yao-02a/"
os.listdir(run_path)
```

```python
run  = Run(run_path)
```

```python
time = 0.0
```

```python
N = run.GetNi(time)
```

```python
fig, ax = plt.subplots(figsize=(10,4))

N.plot(qty='value', \
       ax=ax, \
       vmin=0.2, \
       vmax=1.0, \
       cmap='viridis', \
       ylabel='charge density')
```

```python
B = run.GetB(time)
```

```python
fig, ax = plt.subplots(figsize=(10,4))

B.plot(qty='z', \
       ax=ax, \
       vmin=0.9, \
       vmax=1.1, \
       cmap='viridis', \
       ylabel='Magnetic Field')
```

```python
V = run.GetVi(time)
```

```python
fig, ax = plt.subplots(figsize=(10,4))

V.plot(qty='x', \
       ax=ax, \
       # vmin=0.9, \
       # vmax=1.1, \
       cmap='viridis', \
       ylabel='Ion Velocity')
```

```python
v = modulus(V)
```

```python
fig, ax = plt.subplots(figsize=(10,4))

v.plot(qty='value', \
       ax=ax, \
       # vmin=0.9, \
       # vmax=1.1, \
       cmap='viridis', \
       ylabel='Ion Velocity')
```

```python
v_ = sqrt(dot(V, V))
```

```python
fig, ax = plt.subplots(figsize=(10,4))

v_.plot(qty='value', \
       ax=ax, \
       # vmin=0.9, \
       # vmax=1.1, \
       cmap='viridis', \
       ylabel='Ion Velocity')
```

```python
ions = run.GetParticles(time, ["main"])
```
